```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
all_data = pd.read_csv("all_sales_data.csv")
all_data.head()
```

```python
# Cleaning up the data. (Removing NaN's)
```

```python
# For locating the NaN data.
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

# For removing the NaN data.
all_data = all_data.dropna(how="all")
all_data.head()
```

```python
# For removing data that isn't filled in.
# While cleaning and trying to get all of the data that is organized by month we found data that isn't properly filled in.
all_data = all_data[all_data["Order Date"].str[0:2] != "Or"]
all_data.head()
```

```python
# Adding additional columns to the data
```

```python
all_data["Month"] = all_data["Order Date"].str[0:2]
all_data["Month"] = all_data["Month"].astype("float64")
all_data.head()
```

```python
# Changing needed data to numeric.
```

```python
all_data["Price Each"] = pd.to_numeric(all_data["Price Each"])
all_data["Quantity Ordered"] = pd.to_numeric(all_data["Quantity Ordered"])
```

```python
# Adding a sales column
```

```python
all_data["Sales"] = all_data["Quantity Ordered"] * all_data["Price Each"]
all_data.head()
```

```python
# Totalling our Sales and grouping it by months.
```

```python
results = all_data.groupby("Month").sum()
results
```

```python
# Barchart for showing Sales by Month
months = range(1,13)
plt.bar(months, results["Sales"])
plt.xticks(months)
plt.xlabel("Month number")
plt.ylabel("Sales in USD ($)")
plt.show()
```

```python
# Using the .apply() function to grab our key city values between the comma delimiters.
def get_city(address):
    return address.split(",")[1]

def get_state(address):
    return address.split(",")[2].split(" ")[1]

all_data["City"] = all_data["Purchase Address"].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

```python
results = all_data.groupby("City").sum()
results
```

```python
# Barchart for showing Sales by City
plt.ticklabel_format(axis='both', style='sci', scilimits=(0,10))
cities = [city for city, df in all_data.groupby("City")]
plt.bar(cities, results["Sales"])
plt.xticks(cities, rotation="vertical", size=8)
plt.xlabel("City")
plt.ylabel("Sales in USD ($)")
plt.show()
```

```python
# Now we're going to figure out the best timeframe to display advertisements to maximize the probability of a customer buying a product.
```

```python
# Getting Hour and Minute data.
all_data["Order Date"] = pd.to_datetime(all_data["Order Date"])
```

```python
all_data["Hour"] = all_data["Order Date"].dt.hour
all_data["Minute"] = all_data["Order Date"].dt.minute
all_data["Count"] = 1
all_data.head()
```

```python
keys = [pair for pair, df in all_data.groupby(['Hour'])]

plt.plot(keys, all_data.groupby(['Hour']).count()['Count'])
plt.xticks(keys)
plt.grid()
plt.show()

# Target ads at 11 am - 12 am or 7 pm.
```

```python
# Now we're figuring out what products are most often sold together.
df = all_data[all_data["Order ID"].duplicated(keep=False)]

df["Grouped"] = df.groupby("Order ID")["Product"].transform(lambda x: ",".join(x))

df = df[["Order ID", "Grouped"]].drop_duplicates()

df.head()
```

```python
# Making a list without duplicates filled with the purchase combinations ordered by frequency.
from itertools import combinations
from collections import Counter

count = Counter()

for row in df["Grouped"]:
    row_list = row.split(",")
    count.update(Counter(combinations(row_list, 2))) # Can change the quantity of items within the combinations.

for key, value in count.most_common(10):
    print(key, value)
```

```python
product_group = all_data.groupby("Product")
quantity_ordered = product_group.sum()["Quantity Ordered"]

products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.xlabel("Products")
plt.ylabel("Quantity Ordered")
plt.xticks(products, rotation="vertical", size=8)
plt.show()
```

```python
prices = all_data.groupby("Product").mean()["Price Each"]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color = "g")
ax2.plot(products, prices)

ax1.set_xlabel("Product Name")
ax1.set_ylabel("Quantity Ordered", color = "g")
ax2.set_ylabel("Price ($)", color = "b")
ax1.set_xticklabels(products, rotation = "vertical", size=8)

fig.show()
```
